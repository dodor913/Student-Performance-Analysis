# **Case Study: Student Performance Analytics**

```python
#
import pandas as pd

student_data = {
    'Student': ['Ankush', 'Priya', 'Vivek', 'Sarah', 'Sohan', 'Rina', 'Akash', 'Jyoti', 'Ravi', 'Millie'],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F'],
    'Age': [15, 16, 14, 15, 17, 14, 16, 17, 15, 16],
    'Grade': [9, 10, 8, 9, 11, 8, 10, 11, 9, 10],
    'Attendance (%)': [95, 88, 99, 90, 89, 97, 92, 85, 93, 98],
    'Hours Studied': [3.5, 4.2, 2.1, 3.8, 4.9, 1.7, 3.2, 4.5, 2.7, 5.0],
    'Exam Score': [88, 78, 92, 76, 91, 80, 84, 87, 79, 95],
    'School': ['A', 'B', 'A', 'C', 'D', 'B', 'A', 'D', 'C', 'B'],
    'Extracurricular': ['Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes']
}
df2 = pd.DataFrame(student_data)
```

```python
df2
```

**Insights/Visualizations**
- Performance trends by grade or age: Line plot
- Bar charts: Compare average exam scores by gender or school
- Bar charts: Which school has highest average score?
- Pie chart: Proportion of students in extracurricular activities
- Boxplot: Exam scores split by gender or grade
- Boxplots/Histogram: Distribution of scores or hours studied
- Histogram: Attendance distribution
- Scatter plots: Hours Studied vs Exam Score, Attendance vs Exam Score

```python
import matplotlib.pyplot as plt
```

```python
grade_score = df2.groupby('Grade')['Exam Score'].mean()
```

```python
gender_score = df2.groupby('Gender')['Exam Score'].mean()
gender_score
```

```python
fig, ax = plt.subplots()
ax.plot(grade_score.index, grade_score.values,
        linestyle = '--', color = 'red',
        marker = 'o', label = 'Average score')
ax.set_title('Average Exam Score by Grade')
ax.set_xlabel('Grade')
ax.set_ylabel('Avg Score')
ax.grid(True)
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(gender_score.index, gender_score.values)
ax.set_title('Average Exam Score by Gender')
ax.set_xlabel('Grade')
ax.set_ylabel('Avg Score')
ax.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.boxplot(df2['Hours Studied'])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(df2['Exam Score'], df2['Hours Studied'])
plt.show()
```
